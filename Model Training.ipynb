```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout,Input,Flatten,Dense,MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
batchsize=8
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,rotation_range=0.2,shear_range=0.2,
            zoom_range=0.2,width_shift_range=0.2,height_shift_range=0.2,validation_split=0.2)
```

```python
train_data=train_datagen.flow_from_directory(r'D:\7th_sem\Driver_Drowsiness\Prepared_Data\train',
                                            target_size=(80,80),batch_size=8 ,class_mode ='categorical',subset='training')

validation_data=train_datagen.flow_from_directory(r'D:\7th_sem\Driver_Drowsiness\Prepared_Data\train',
                                            target_size=(80,80),batch_size=8 ,class_mode ='categorical',subset='validation')
```

```python
test_datagen=ImageDataGenerator(rescale=1/.255)

test_data = test_datagen.flow_from_directory(r'D:\7th_sem\Driver_Drowsiness\Prepared_Data\test',
                                            target_size=(80,80),batch_size=8 ,class_mode ='categorical',subset='training')
```

```python
#model

basemodel = InceptionV3(include_top=False, weights= 'imagenet',input_tensor = Input(shape=(80,80,3)))
hmodel =basemodel.output
hmodel = Flatten()(hmodel)
hmodel = Dense(64,activation = 'relu')(hmodel)

hmodel = Dropout(0.5)(hmodel)
hmodel = Dense(2,activation = 'softmax')(hmodel)

model = Model(inputs=basemodel.input, outputs=hmodel)

for layer in basemodel.layers:
    layer.trainable = False 
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
```

```python
checkpoint = ModelCheckpoint(r'D:\7th_sem\Driver_Drowsiness\model',monitor ='val_loss',save_best_only=True ,verbose =3)

earlystop = EarlyStopping(monitor ='val_loss',patience = 7 ,verbose =3 , restore_best_weights= True)

learning_rate = ReduceLROnPlateau(monitor =' val_loss',patience =3 , verbose =3)

callbacks=[checkpoint,earlystop,learning_rate]
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(train_data,
          steps_per_epoch=train_data.samples // batchsize,
          validation_data=validation_data,
          validation_steps=validation_data.samples // batchsize,
          callbacks=callbacks,
          epochs=15)
```
